###My workflow, data validity 

###1
    
    I do research in the field of biomineralization. Nano crystals that are
    present for example in bone material. 
    
    
	The data that I use is generated by model fitting with measured data.
    
	The results of calculation are spread across multiple files, with diverse
    internal structure.
    I measure lot of samples, and for each I need to make a model fit. 
    So my database is a collection of calculation results for all the samples.
    
    Each fit is with 20-40 parameters. 
    
    My main work is to play with the models as crystallographic structure models,
    where the variation of the parameters signify different description of physical
    properties of the material.
    
    In my case, the data analysis is an overlayer, a different perspective, 
    an overview on the behaviour and interplay of the parameter. 
    There are lot of correlation from sample to sample above (next to) the correlation 
    of individual parameter inside the model based fit.
    
    Using linux bash scripts and python-pandas:
        I monitor my calculations,
        identify values that are off, 
        recognizing correlations (patterns)
        identifying outliers.
